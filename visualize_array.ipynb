```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
paths = ['weights_zeors.npz', 'weights_ones.npz', 'weights_masked.npz',
         'weights_no_action_mask1.npz', 'weights_added.npz', 'weights_clear_pred_mask.npz',
         'weights_no_action_mask2.npz']

for id, weight_path in enumerate(paths):
    weights = np.load(weight_path)
    plt.figure(figsize=(10,20))
    plt.subplot(7,1,id+1)
    plt.plot(weights['a'][0:100], color='red')
    plt.ylim(-0.2,1.2)
    plt.title(weight_path)

plt.subplots_adjust(hspace=30)
plt.show()
```
